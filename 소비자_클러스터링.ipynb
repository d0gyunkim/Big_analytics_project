```python
# 라이브러리 호출
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# 그래프상에서 한글을 나타내기 위해 korean_mat 설치
! pip install koreanize-matplotlib
import koreanize_matplotlib

%config InlineBackend.figure_format = 'retina'

# 데이터 불러오기
df = pd.read_csv('/Users/gimdogyun/Desktop/빅데이터_애널리틱스_프로젝트/Big_analytics_project/movie_finished_toFDA_V9.csv')
df.head()
```

```python
# 2주_누적관객수
# Group by movie title and calculate the cumulative audience count (sum)
df['2주_누적관객수'] = df.groupby('movie_title')['관객수'].transform('sum')
```

```python
col = [
    'running_time', 'watch_grade',
    '사극', 'SF', '어드벤처', '액션', '범죄', '판타지', '가족', '미스터리', '스릴러', '뮤지컬',
    '코미디', '공포(호러)', '애니메이션', '드라마', '서부극(웨스턴)', '전쟁', '멜로/로맨스', '기타',
    '다큐멘터리', '공연', '성인물(에로)',
    'NTZ_male_score', 'NTZ_female_score', 'NTZ_teen_score', 'NTZ_twenties_score', 
    'NTZ_thirties_score', 'NTZ_forties_score', 'NTZ_fifties_more_score',
    'NTZ_direct_point', 'NTZ_act_point', 'NTZ_story_point', 
    'NTZ_visual_point', 'NTZ_ost_point',
    'director_power', 'actor_power',
    'RV_male', 'RV_female', 'RV_teen', 'RV_twenties', 
    'RV_thirties', 'RV_forties', 'RV_fifties','movie_title'
]
```

```python
pd.set_option('display.max_columns', None)
df_clustering = df[col].copy()
```

```python
# log 변환
df_clustering['director_power'] = np.log1p(df_clustering['director_power'])
df_clustering['actor_power'] = np.log1p(df_clustering['actor_power'])
```

```python
# 비율 피쳐: 30% -> 0.3
ratio_col = ['NTZ_direct_point','NTZ_act_point','NTZ_story_point','NTZ_visual_point','NTZ_ost_point', 'RV_male','RV_female','RV_teen','RV_twenties','RV_thirties','RV_forties','RV_fifties']

# 비율 변환
for col in ratio_col:
    df_clustering[col] = df_clustering[col].str.rstrip('%').astype(float) / 100
```

```python
df_clustering
```

```python
# min-max scaling
from sklearn.preprocessing import MinMaxScaler

# min-max를 사용해야 할 칼럼: 
min_max_col = ['running_time', 'NTZ_male_score', 'NTZ_female_score', 'NTZ_teen_score', 'NTZ_twenties_score','NTZ_thirties_score', 'NTZ_forties_score', 'NTZ_fifties_more_score', 'director_power','actor_power']

scaler = MinMaxScaler()
for col in min_max_col:
    df_clustering[col] = scaler.fit_transform(df_clustering[[col]])
```

```python
# watch_grade 더미화
df_clustering = pd.get_dummies(df_clustering, columns=['watch_grade'])
# True, False -> 1, 0: watch_grade
df_clustering.update(df_clustering.select_dtypes(include=['bool']).astype(int))
```

```python
# movie_title로 groupby 후 평균값으로 대체
df_clustering = df_clustering.groupby('movie_title').mean()
```

```python
# k-means clustering
from sklearn.cluster import KMeans

# kmeans 모델 생성
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(df_clustering)

# 군집화 결과 확인
df_clustering['cluster'] = kmeans.labels_
df_clustering['cluster'].value_counts()

# 각 군집별로 평균값
cluster_mean = df_clustering.groupby('cluster').mean()
cluster_mean
```

```python
# 원래 df에 군집 결과를 합침
df = pd.merge(df, df_clustering['cluster'], on='movie_title', how='left')
```
